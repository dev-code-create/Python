```python
from sklearn.linear_model import LinearRegression

X = [[1],[2],[3],[4],[5]]
y = [40,50,65,75,90]

model = LinearRegression()
#model training

model.fit(X,y)

Hours = float(input("Enter How many hours"))

predicted_value = model.predict([[Hours]])

print(f"based on your hours {Hours} your marks will be {predicted_value}")
```
