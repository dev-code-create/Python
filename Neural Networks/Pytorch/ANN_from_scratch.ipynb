```python
import torch
import torch.nn as nn

class ANN(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.linear(2,16)
    self.fc2 = nn.linear(16,8)
    self.out = nn.linear(8,1)

  def forward(self,x):
    x = torch.relu(self.fc1(x))
    x = torch.relu(self.fc2(x))
    x = self.out(x)
    return x
```
